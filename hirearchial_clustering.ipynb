```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('/Users/aditirajesh/Downloads/Mall_Customers.csv')
```

```python
X = data[['Annual Income (k$)', 'Spending Score (1-100)']].values
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
linked = linkage(X_scaled, method='ward')
```

```python
plt.figure(figsize=(10, 7))
dendrogram(linked)
plt.axhline(y=8, color='r', linestyle='--')  # Adjust 'y' to an appropriate threshold based on the dendrogram
plt.title('Dendrogram for Mall Customers')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distance')
plt.show()
```

```python
# Plot the dendrogram
plt.figure(figsize=(10, 7))
plt.title("Dendrogram for Mall Customers")
dendrogram(linked, truncate_mode='level', p=12)  # Truncate to show main branches clearly

# Adjust the threshold line
plt.axhline(y=3.8, color='r', linestyle='--')  # Set threshold slightly below 4
plt.xlabel("Customers")
plt.ylabel("Euclidean Distance")
plt.show()
```
